```python
baseline = results["plateaus"]["baseline"]
adaptive = results["plateaus"]["ipopt_adaptive"]

print("\nBaseline plateau:")
print(baseline)

print("\nIPOPT Adaptive plateau:")
print(adaptive)

if baseline["found"] and adaptive["found"]:
    baseline_level = baseline["plateau_level"]
    adaptive_level = adaptive["plateau_level"]

    print("\nBaseline plateau level:", baseline_level)
    print("Adaptive plateau level:", adaptive_level)
    print("Baseline / Adaptive:", baseline_level / adaptive_level)

    if baseline_level > adaptive_level:
        print("现象成立：Baseline plateau 高于 Adaptive。")
    else:
        print("现象不成立：Baseline plateau 没有高于 Adaptive。")
else:
    print("至少一边没有检测到 plateau，需要增加预算或调整检测参数.")

print("Gradient-evaluation plot:",
      results["plots"]["baseline_vs_ipopt_gradient_evaluations"])
print("CPU-time plot:",
      results["plots"]["baseline_vs_ipopt_cpu_time"])
```
